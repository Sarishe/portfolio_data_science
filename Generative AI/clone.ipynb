```python
import sys
sys.path.append("/content/Real-Time-Voice-Cloning")

from pathlib import Path
from synthesizer.inference import Synthesizer
from encoder import inference as encoder
from vocoder import inference as vocoder

encoder_model_path = Path('/content/drive/MyDrive/Colab Notebooks/Generative AI/saved_models/default/encoder.pt')
synthesizer_model_path = Path('/content/drive/MyDrive/Colab Notebooks/Generative AI/saved_models/default/synthesizer.pt')
vocoder_model_path = Path('/content/drive/MyDrive/Colab Notebooks/Generative AI/saved_models/default/vocoder.pt')

encoder.load_model(encoder_model_path)
synthesizer = Synthesizer(synthesizer_model_path)
vocoder.load_model(vocoder_model_path)
```

```python
import numpy as np
import librosa
import soundfile as sf
import IPython.display as ipd

input_wav = "/content/drive/MyDrive/Colab Notebooks/Generative AI/recorded.wav"
original_wav, sampling_rate = librosa.load(input_wav, sr=16000)

embed = encoder.embed_utterance(original_wav)

text = "Heeeey, this is a cloned voice! Please, take your seat near the window"

specs = synthesizer.synthesize_spectrograms([text], [embed])
generated_wav = vocoder.infer_waveform(specs[0])

output_path = "/content/cloned_voice.wav"
sf.write(output_path, generated_wav, 16000)

ipd.Audio(output_path)
```

```python
import numpy as np
import librosa
import soundfile as sf
import IPython.display as ipd

input_wav = "/content/drive/MyDrive/Colab Notebooks/Generative AI/recorded_1.wav"
original_wav, sampling_rate = librosa.load(input_wav, sr=16000)

embed = encoder.embed_utterance(original_wav)

text = "Heeeey, this is a cloned voice! Please, take your seat near the window"

specs = synthesizer.synthesize_spectrograms([text], [embed])
generated_wav = vocoder.infer_waveform(specs[0], normalize=True)

output_path = "/content/cloned_voice.wav"
sf.write(output_path, generated_wav, 16000)

ipd.Audio(output_path)
```

```python
from bark import generate_audio, SAMPLE_RATE
from scipy.io.wavfile import write as write_wav
import IPython.display as ipd
import numpy as np
import textwrap
import docx
from google.colab import drive
import spacy
import time

nlp = spacy.load("en_core_web_sm")

drive.mount('/content/drive')

def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = ""
    for para in doc.paragraphs:
        full_text += para.text + "\n"
    return full_text

docx_file_path = '/content/drive/MyDrive/Colab Notebooks/Generative AI/text.docx'
text_prompt = read_docx(docx_file_path)


def split_into_sentences(text):
    doc = nlp(text)
    return [sent.text.strip() for sent in doc.sents]

sentences = split_into_sentences(text_prompt)

max_chunk_length = 200
chunks = []
current_chunk = ""

for sentence in sentences:
    if len(current_chunk) + len(sentence) <= max_chunk_length:
        current_chunk += " " + sentence
    else:
        chunks.append(current_chunk.strip())
        current_chunk = sentence

if current_chunk:
    chunks.append(current_chunk.strip())

voice = "v2/en_speaker_0"

audio_segments = []

for i, chunk in enumerate(chunks):
    print(f"\nChunk {i+1}/{len(chunks)}:\n{chunk}\n")
    audio = generate_audio(chunk, history_prompt=voice)
    audio_segments.append(audio)
    time.sleep(0.5)

full_audio = np.concatenate(audio_segments)

write_wav("full_story.wav", SAMPLE_RATE, full_audio)
ipd.Audio("full_story.wav")
```

```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect('database.db')
```

```python
sql_query1 = '''
CREATE TABLE table1 (
    id TEXT PRIMARY KEY,
    vehicle_type TEXT,
    milleage INTEGER,
    max_speed INTEGER,
    passenger_count INTEGER,
    category TEXT
    )'''
```

```python
cursor = connection.cursor()
try:
    cursor.execute(sql_query1) 
    connection.commit() 
except Exception as e:
    print(e)
    connection.rollback()
```

```python
#Таблиця, яка містить Водіїв, що можуть їздити лише по двох типах доріг. 
select = '''
SELECT *
FROM table2
JOIN table1 ON table1.id = table2.car_id 
JOIN table3 ON table3.category = table1.category
WHERE (table1.category = table2.driver_category) AND ((permitted_roads='AC34') OR (permitted_roads = 'MR55'))'''
```

```python
pd.read_sql_query(select, connection)
```
